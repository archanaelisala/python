```python
import matplotlib.pyplot as plt
x = [7,4,3,8,2,5]
y = [10,24,23,7,3,6]
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [7,4,3,8,2,5]
y = [10,24,23,7,3,6]
plt.bar(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [10,24,23,7,3,6]
plt.hist(y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
